# Outlier Removal

```python
datates['TERRITORY'].value_counts()
```

```python
import numpy as np
```

Remove outlier QUANTITYORDERED

```python
Q1 = np.percentile(datates['QUANTITYORDERED'], 15,
                   interpolation = 'midpoint')
Q3 = np.percentile(datates['QUANTITYORDERED'], 65,
                   interpolation = 'midpoint')
IQR = Q3 - Q1
print("Old Shape : ", datates.shape)
```

```python
# Upper bound
upper = np.where(datates['QUANTITYORDERED'] >= (Q3+1.5*IQR))
# Lower bound
lower = np.where(datates['QUANTITYORDERED'] <= (Q1-1.5*IQR))

''' Removing the Outliers '''
datates.drop(upper[0], inplace = True)
datates.drop(lower[0], inplace = True)

print("New Shape: ", datates.shape)
```

Remove outlier SALES

```python
Q1 = np.percentile(datates['SALES'], 10,
                   interpolation = 'midpoint')
Q3 = np.percentile(datates['SALES'], 75,
                   interpolation = 'midpoint')
IQR = Q3 - Q1
print("Old Shape : ", datates.shape)
```

```python
# Upper bound
upper = np.where(datates['SALES'] >= (Q3+1.5*IQR))
# Lower bound
lower = np.where(datates['SALES'] <= (Q1-1.5*IQR))

''' Removing the Outliers '''
datates.drop(upper[0], inplace = True)
datates.drop(lower[0], inplace = True)

print("New Shape: ", datates.shape)
```

Remove Outlier MSRP

```python
Q1 = np.percentile(datates['MSRP'], 25,
                   interpolation = 'midpoint')
Q3 = np.percentile(datates['MSRP'], 75,
                   interpolation = 'midpoint')
IQR = Q3 - Q1
print("Old Shape : ", datates.shape)
```

```python
# Upper bound
upper = np.where(datates['MSRP'] >= (Q3+1.5*IQR))
# Lower bound
lower = np.where(datates['MSRP'] <= (Q1-1.5*IQR))

''' Removing the Outliers '''
datates.drop(upper[0], inplace = True)
datates.drop(lower[0], inplace = True)

print("New Shape: ", datates.shape)
```

Logistic Regression

```python
# Menggunakan data frame untuk membuat variabel dummy
#y = df_dummies['Churn'].values
y = y
X = datates

# Scaling all the variables to a range of 0 to 1
from sklearn.preprocessing import MinMaxScaler
features = X.columns.values
scaler = MinMaxScaler(feature_range = (0,1))
scaler.fit(X)
X = pd.DataFrame(scaler.transform(X))
X.columns = features
```

```python
# Create Train & Test Data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
# Running logistic regression model
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
result = model.fit(X_train, y_train)
```

```python
from sklearn import metrics
prediction_test = model.predict(X_test)
# Print the prediction accuracy
print (metrics.accuracy_score(y_test, prediction_test))
```

```python
# getting bobot dari semua variabel
weights = pd.Series(model.coef_[0],
                 index=X.columns.values)
print (weights.sort_values(ascending = False)[:10].plot(kind='bar'))
```

```python
print(weights.sort_values(ascending = False)[-10:].plot(kind='bar'))
```

Beberapa variable memiliki hubungan negatif pada variable TERRITORY dan lainnya memiliki hubungan positif

Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)
model_rf = RandomForestClassifier(n_estimators=1000 , oob_score = True, n_jobs = -1,
                                  random_state =50, max_features = "auto",
                                  max_leaf_nodes = 30)
model_rf.fit(X_train, y_train)

# Make predictions
prediction_test = model_rf.predict(X_test)
print (metrics.accuracy_score(y_test, prediction_test))
```

```python
importances = model_rf.feature_importances_
weights = pd.Series(importances,
                 index=X.columns.values)
weights.sort_values()[-10:].plot(kind = 'barh')
```

Support Vecor Machine (SVM)

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=99)
```

```python
from sklearn.svm import SVC

model.svm = SVC(kernel='linear')
model.svm.fit(X_train,y_train)
preds = model.svm.predict(X_test)
metrics.accuracy_score(y_test, preds)
```

```python
# Confusion matrix
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test,preds))
```
